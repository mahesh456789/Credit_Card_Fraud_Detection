```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset into pandas dataframe 
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information 
credit_card_data.info()
```

```python
#checking the no.of missing values in each column 
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transactions 
credit_card_data['Class'].value_counts()
```

```python
#this dataset is highly unbalanced 0-normal transcations 1-fraudulent transactions
```

```python
#seprating the data for analysis 
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions  
credit_card_data.groupby('Class').mean()
```

```python
#Under Sampling 
#build a sample dataset containing similar distribution of narmal and fraudulent transactions 
#no.of fraudulent transactions is 492 
#when data is balanced it is easy for machine learning algorithms to predict.so,take 492 from normal transactions combine or join with fraudulent transactions
```

```python
#it will take random legit or normal transactions called random sampling
legit_sample=legit.sample(492)
```

```python
legit_sample
```

```python
#concatenating two data frames 
#axis=0 means added by rows 
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#spliting the data into features and targets
#features are v1,v2...... targets are class values 0 or 1 store in variables
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
#split the data into training data and testing data 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model training (Logistic Regression)
model=LogisticRegression()
```

```python
#training the Logistic regression model with training data 
model.fit(X_train,Y_train)
```

```python
#model Evaluation 
#accuracy on training data 
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
#Accuracy score on training data 
training_data_accuracy
```

```python
#accuracy on test data 
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
#Accuracy score on testing data 
testing_data_accuracy
```
